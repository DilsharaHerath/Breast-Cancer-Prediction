# Data Collection and Preprocessing

Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets as sd
from sklearn.model_selection import train_test_split
```

Data Collection and Preprocessing

```python
breast_cancer_dataset = sd.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# Adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

Seperating the features and targets

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)
# Fitting ther data should be only done to the training data

x_test_std = scaler.transform(x_test)
```

```python
print(x_train_std)
```

# Building the Neural Network

```python
# Importing  tensorflow and keras
import tensorflow as tf   # Deep Learning library developed by Google
tf.random.set_seed(3)     # No matter you run the NN so many times, you will get the same accuracy.
# Otherwise weights and parameters will be different when running again and again
from tensorflow import keras  # uses tf as backend and make the development of NN more easier
```

Setting up the layers of the NN

```python
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(2,activation='sigmoid')
])
```

**Flatten** : We need to flatten the input matrix before edding to the NN. Each feature should go to one input neuron. Therefore, #input_neurons=#fetures

The use of flatten is to convert the data into a single dimensional array.

## Compiling the NN - For non-standardized data

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

When integer values are used o categorical values **Sparse_categorical_crossentropy** is used


dog - 0
cat - 1
horse - 2

When one hot encoding is used we will use **categorical_crossentropy** only.


dog - [1 0 0]
cat - [0 1 0]
horse - [0 0 1]


## Training the NN


```python
history = model.fit(x_train, y_train, validation_split=0.1, epochs=10)
```

**Ephocs** : how many times you need to go through the data

This is getting low accuracies compared to the ML models created for the same dataset. This is due to the data has not being standardiazed.

## Compiling the NN - For Standardized data

```python
history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=10)
```

## Visualizing the accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Loss Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```

Building the predictive system
